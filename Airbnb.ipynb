```python
import pandas as pd

# Create lists to store data
unique_listing_id = []
listing_title = []
listing_description = []
unique_host_id = []
host_name = []
neighbourhood_name = []
coordinates = []
listing_price = []
start_date = []
end_date = []
amenities = []
average_rating = []
reviews_data = []

print(all_documents)
# Iterate through the fetched data and populate lists
for document in all_documents:
    unique_listing_id.append(document.get("_id"))
    listing_title.append(document.get("name"))
    listing_description.append(document.get("description"))
    unique_host_id.append(document.get("host_id"))
    host_name.append(document.get("host_name"))
    neighbourhood_name.append(document.get("neighbourhood"))
    coordinates.append(document.get("location", {}).get("coordinates"))
    listing_price.append(document.get("price"))
    start_date.append(document.get("availability", {}).get("start_date"))
    end_date.append(document.get("availability", {}).get("end_date"))
    amenities.append(document.get("amenities"))
    average_rating.append(document.get("rating"))
    reviews_data.append(document.get("reviews"))

# Create a dictionary from the lists
data_dict = {
    "_id": unique_listing_id,
    "name": listing_title,
    "description": listing_description,
    "host_id": unique_host_id,
    "host_name": host_name,
    "neighbourhood": neighbourhood_name,
    "coordinates": coordinates,
    "price": listing_price,
    "start_date": start_date,
    "end_date": end_date,
    "amenities": amenities,
    "rating": average_rating,
    "reviews": reviews_data
}

# Create a DataFrame from the dictionary
df = pd.DataFrame(data_dict)

# Close the MongoDB connection
client.close()

```
